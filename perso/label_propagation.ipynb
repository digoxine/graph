```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.vq import kmeans,vq
import random
```

```python
G = nx.Graph()
```

```python
f = open('jaccard_label.txt', 'r')
cpt = 0
for line in f:
    lis = line.split() 
    G.add_node((int) (cpt),group=(int)(lis[0]))
    cpt+=1
f.close()
```

```python
#Attribution de pos
position = []
labels = []
MAX_X = 1500
MAX_Y = 1500
NB_CLUSTERS = 6
NB_NODES = 400
cpt = 0
NB_X = NB_CLUSTERS//2
NB_Y = NB_CLUSTERS - NB_X
GRILLE_X = (MAX_X) //(NB_X)
GRILLE_Y = (MAX_Y) // NB_Y
f = open('jaccard_label.txt', 'r')
for line in f:
    current_cluster = (int) ( cpt/ (NB_NODES/NB_CLUSTERS) )
    labels.append(current_cluster)
    current_nb_x = (current_cluster//NB_X)
    current_nb_y = (current_cluster%NB_X)
    x = random.randint((current_nb_x)*GRILLE_X,(current_nb_x+1)*GRILLE_X)
    y = random.randint((current_nb_y)*GRILLE_Y,(current_nb_y+1)*GRILLE_Y)
    position.append((x,y))
    cpt+=1
f.close()
```

```python
labels = {}
labels[0] = r'$a$'
labels[1] = r'$b$'
labels[2] = r'$c$'
labels[3] = r'$d$'
labels[4] = r'$\alpha$'
labels[5] = r'$\beta$'
labels[6] = r'$\gamma$'
labels[7] = r'$\delta$'
plt.plot()
nx.draw_networkx_labels(G, position, labels, font_size=16)
plt.axis('off')
plt.show()
```

```python
from itertools import count
groups = set(nx.get_node_attributes(G, 'group').values())
mapping = dict(zip(sorted(groups),count()))
nodes = G.nodes()
colors = [mapping[G.nodes[n]['group']] for n in nodes]
nc = nx.draw_networkx_nodes(G,position,nodelist=nodes, node_color=colors, with_labels=False, node_size=100, cmap=plt.cm.jet)
plt.colorbar(nc)
plt.axis('off')
plt.show()
```

```python
from networkx.drawing.nx_pydot import write_dot
pos = nx.nx_agraph.graphviz_layout(G)
nx.draw(G, pos=position)
```
